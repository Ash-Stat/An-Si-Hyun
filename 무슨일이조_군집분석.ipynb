이전 요인분석에서 영업활성화 요인의 경우는 주로 손익계산서 계정을 이루는 항목으로 구성이 되어있다. 따라서 건실한 손익계정을 뜻하는 **'Income'**으로 명명한다.

# 필요한 모듈 다운로드

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# 이전 요인분석 점수에 대한 데이터셋 다운로드, 그리고 칼럼명 변경

```python
cl_data=pd.read_csv('./fa_data.csv')
cl_data.columns=['Income','Debt','Operating profit','Capital']
cl_data
```

# 기초통계량 조사

전처리(표준화 스케일링, 결측값 제거 등)는 요인 분석시에 다 실시 했으므로, 따로 실시하지 않는다.

```python
cl_data.describe()
```

# hier- 계층 군집분석을 위한 데이터셋
# kmean- k-mean 군집분석을 위한 데이터셋

```python
hier=cl_data
kmean=cl_data
```

# 계층 군집분석 실시
single, complete, average, centroid, ward 5 가지 방법을 실시하며 결과를 비교해 본다.

```python
from scipy.cluster.hierarchy import linkage, dendrogram
cluster_single=linkage(y=hier, method='single',metric='euclidean')
cluster_complete=linkage(y=hier, method='complete',metric='euclidean')
cluster_average=linkage(y=hier, method='average',metric='euclidean')
cluster_centroid=linkage(y=hier, method='centroid',metric='euclidean')
cluster_ward=linkage(y=hier, method='ward',metric='euclidean')
```

```python
pd.DataFrame(cluster_single)
```

# centroid 메소드

```python
pd.DataFrame(cluster_centroid)
```

# ward 메소드

```python
pd.DataFrame(cluster_ward)
```

# 덴드로그램 시각화 결과

층이 높아질 수록 군집들이 뚜렷하게 나누어지는 방법이라고 할 수 있다.
다시 말해서, 다른 방법들에 비해서 ward의 경우는 20층에 가서야 하나의 군집으로 모이게 되었는데 이는 ward 방법이 위 데이터셋에서 다른 방법들에 비해 군집들을 잘 나누는 방법이 될 것이라고 짐작해 볼 수 있다.

# 군집의 갯수를 몇개로 나눌까?

요인 분석의 결과 4개의 요인들로 압축할 수 있었다.
군집이 많아질 수록 해석이 어려우므로 이에 군집들을 4개의 군집으로 요약해 보기로 한다.

# 군집을 네 개로 나눌 수 있는 t값 찾기

## single


```python
from scipy.cluster.hierarchy import fcluster
cut_single=fcluster(cluster_single, t=5.5, criterion='distance')
cut_single.max()
```

## complete


```python
cut_complete=fcluster(cluster_complete, t=10.5, criterion='distance')
cut_complete.max()
```

## average


```python
cut_average=fcluster(cluster_average, t=8, criterion='distance')
cut_average.max()
```

## centroid

```python
cut_centroid=fcluster(cluster_centroid, t=8, criterion='distance')
cut_centroid.max()
```

## ward

```python
cut_ward=fcluster(cluster_ward, t=17, criterion='distance')
cut_ward.max()
```

# 각각 방법들로 나눈 군집들을 데이터 셋에 합쳐준다.

```python
hier['single']=cut_single
hier['complete']=cut_complete
hier['average']=cut_average
hier['centroid']=cut_centroid
hier['ward']=cut_ward
hier
```

# 방법별 실루엣 점수 결과
좋은 군집분석이 되기 위해서는 같은 군집내끼리는 동질성을 지니고, 다른 군집간에는 이질성이 있어야한다.

실루엣 계수로 이를 해석해보자면, 1에 가까울수록 군집간 이질성이 높고, 0에 가까울수록 군집간 이질성이 약하다는 뜻이다. 또한 -계수는 군집내 동질성이 떨어짐을 뜻한다.

이러한 면에서 보았을때, ward의 방법은 다른 군집과 마찬가지로 - 계수를 포함하고 있기는 하나 그 정도가 제일 약하고, + 계수를 갖는 군집은 그 점수가 다른 방법들에 비해 높은 편이다.

이를 정리하면, ward의 방법은 비교적 다른 방법들에 비해 군집들간의 내부 동질성을 높게, 군집들간의 외부 이질성을 뚜렷하게 만들어 준다는 것을 알 수 있다.

 계층 군집분석 정리

단일 연결법(single), 완전 연결법(complete), 평균 연결법(average), 중심 연결법(centroid) 은 모두 유클리드 거리를 기반으로 하는 군집분석이다.

반면에 ward의 경우는 군집간 유사성에 대해서 SSE(오차 제곱합)의 증가분에 기반해서 측정한다.

위의 덴드로그램, 분포, 박스 플랏, 산점도 시각화, 실루엣 점수 등의 결과로도 알 수 있듯이, 오차제곱합의 증가분에 기반하여 군집을 측정하는 ward의 방법이 이 데이터셋을 군집분석하기 위한 가장 적절한 방법임을 일관되게 가리키고 있다.

ward의 방법 결과 각 군집별 특징은 다음과 같다.

## 1번 군집
기초 통계량을 살펴 보았을때 다른 요인들과는 다르게 Debt의 -값이 높게 나왔다. 이는 곧 Debt 즉, 부채의 악화정도가 상대적으로 건실한 병원들이 모여 있다는 뜻이며 이 군집내의 병원들은 부채비율이 상당히 낮을 것으로 추정된다. 따라서 1번 군집을 **"자본비율이 높은 군집"** 으로 명명한다.
자본비율 = 자본/(자본+부채)

## 2번 군집
다른 요인들과는 다르게 Income의 + 값이 제일 뚜렷하게 나왔다. 이는 곧 Income 즉, 손익(수익과 수익을 창출하는데에 있어서 필요한 비용 및 기타 비용등)의 구조가 좋은 병원들이 모여 있다는 뜻이며 이 군집내의 병원들은 손익계산서(Income statement)상에서 비용에 비해 전체적인 수익구조가 높을 것으로 추정된다. 따라서 2번 군집을 **"영업이 활발한 군집"**으로 명명한다.

## 3번 군집
다른 요인들과 비교를 해보았을때 Income의 - 값이 뚜렷하고, Capital 요인의 경우에는 유일한 +값이 나왔다. 이는 곧, 단기지급능력은 있으나 영업 부진함으로 인해서 다소 위험한 병원들이 모여 있다는 뜻이며 이 군집내의 병원들은 재무구조상 폐업의 가능성이 있을 것으로 보인다. 따라서 3번 군집을 **"폐업가능성이 있는 군집"**으로 명명한다.

## 4번 군집
다른 요인들과는 다르게 Capital - 값이 뚜렷하다. 이는 곧, Capital 즉, 자본의 건실함 정도가 약한 병원들이 모여 있다는 뜻이며 이 군집내의 병원들은 단기지급능력이 약한 것으로 추정된다. 따라서 4번 군집을 **"단기지급능력이 약한 군집"**으로 명명한다.

# K-Means 군집분석 실시

```python
# k-meas clustering
kmean=pd.read_csv('./fa_data.csv')
kmean.columns=['Income','Debt','Operating profit','Capital']
kmean
```

# 엘보우(꺾이는점)를 살펴봄으로써 군집의 개수를 결정해 보기 위한 시각화
## 처음으로 꺾이는 지점인 4개의 군집을 선택하기로 한다.

```python
from sklearn.cluster import KMeans
dist=[]
for i in range(1,20):
  km=KMeans(
      n_clusters=i, init='random',
      n_init=10, max_iter=300,
      tol=1e-04, random_state=42
  )
  km.fit(kmean)
  dist.append(km.inertia_)
plt.figure(figsize=(10,10))
plt.plot(range(1,20), dist,marker='*')
plt.xlabel('Number of Clusters')
plt.ylabel('Distortion')
plt.show()
```

# Kmean 군집분석 적합시키기, 시각화 결과로부터 군집수를 5개로 설정한다.

```python
model=KMeans(n_clusters=5, random_state=42,algorithm='auto')
model.fit(kmean)
```

# 군집결과 데이터셋에 합쳐주기

```python
prediction=model.predict(kmean)
kmean['kclusters']=prediction
kmean
```
