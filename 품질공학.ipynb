```
import matplotlib.pyplot as plt
import numpy as np
```

```
def BassModel(t,M,p,q):
  t=t-2020
  a=q/p
  b=p+q
  return M*(1-np.exp(-b*t))/(1+a*np.exp(-b*t))

x=np.arange(2020,2065,0.1)
```

```
plt.title("BassModelRegression M=1,650.041×10^6, q=0.132501")

plt.plot(x, BassModel(x,1650041000,0.002493,0.132501), 'y--', label='p=0.002493')
plt.plot(x, BassModel(x,1650041000,0.0005,0.132501), 'm--', label='p=0.0005')
plt.plot(x, BassModel(x,1650041000,0.0010,0.132501), 'c--', label='p=0.0010')
plt.plot(x, BassModel(x,1650041000,0.0025,0.132501), 'b:', label='p=0.0025')
plt.plot(x, BassModel(x,1650041000,0.0037,0.132501), 'r', label='p=0.0037')
plt.plot(x, BassModel(x,1650041000,0.0050,0.132501), 'b--', label='p=0.0050')
plt.plot(x, BassModel(x,1650041000,0.0075,0.132501), 'g--', label='p=0.0075')
plt.plot(x, BassModel(x,1650041000,0.0100,0.132501), 'k--', label='p=0.0100')


plt.legend(loc='lower right')
plt.xticks([2020,2025,2030,2035,2040,2045,2050,2055,2060,2065])
plt.xlabel("Year")
plt.ylabel("iPhone Sales Volume")
plt.show()
```
