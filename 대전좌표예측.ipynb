```
import pandas as pd
import numpy as np
Daejeon_S1=pd.read_csv('./Daejeon_S1_HGVI.csv')
location=pd.read_csv('./좌표회귀이.txt',sep=' ')
```

```
Daejeon_S1=Daejeon_S1.drop(['Number','계약년','계약월','계약일','도시','구','동','도로명','난방연료','난방연료'],axis=1)
Daejeon_S1
```

```
Daejeon_S1=Daejeon_S1.drop(['아파트 이름'],axis=1)
Daejeon_S1
```

```
location
```

```
Daejeon_S1[['target_west_log','target_east_log','target_north_lat','target_south_lat']]=np.nan
```

```
for v1,v2,v3 in zip(Daejeon_S1['x'],Daejeon_S1['y'],Daejeon_S1.index):
  v1=round(v1,6)
  v2=round(v2,6)
  for c1,c2,c3,c4,c5,c6,c7 in zip(location['train_log'],location['train_lat'],location['target_west_log'],location['target_east_log'],location['target_north_lat'],location['target_south_lat'],location.index):
    c1=round(c1,6)
    c2=round(c2,6)
    c3=round(c3,6)
    c4=round(c4,6)
    c5=round(c5,6)
    c6=round(c6,6)
    if v1 == c1 and v2 == c2:
      Daejeon_S1['target_west_log'].loc[v3] = c3
      Daejeon_S1['target_east_log'].loc[v3] = c4
      Daejeon_S1['target_north_lat'].loc[v3] = c5
      Daejeon_S1['target_south_lat'].loc[v3] = c6
Daejeon_S1
```

```
Daejeon_S1=Daejeon_S1.drop_duplicates()
Daejeon_S1
```

```
target=Daejeon_S1[['target_west_log','target_east_log','target_north_lat','target_south_lat']]
target
```

```
variables=Daejeon_S1.drop(['HGVI','target_west_log','target_east_log','target_north_lat','target_south_lat'],axis=1)
variables
```

```
correlation=Daejeon_S1.corr()
correlation
```

```
from sklearn.model_selection import train_test_split
train,test=train_test_split(Daejeon_S1,test_size=0.2,random_state=42)
```

```
train
```

```
test
```

```
train_x=train.drop(['HGVI','target_west_log','target_east_log','target_north_lat','target_south_lat'],axis=1)
train_y=train[['target_west_log','target_east_log','target_north_lat','target_south_lat']]
test_x=test.drop(['HGVI','target_west_log','target_east_log','target_north_lat','target_south_lat'],axis=1)
test_y=test[['target_west_log','target_east_log','target_north_lat','target_south_lat']]
```

```
for ss,indd in zip(test_x['거래금액(만원)'], test_x.index) :
  ss=ss.replace(',','')
  test_x['거래금액(만원)'].loc[indd] = ss
test_x
```

```
for ss1,indd1 in zip(train_x['거래금액(만원)'], train_x.index) :
  ss1=ss1.replace(',','')
  train_x['거래금액(만원)'].loc[indd1] = ss1
train_x
```

```
train_x
```

```
train_x_x=train_x.replace({'자료없음' : np.nan})
test_x_x=test_x.replace({'자료없음' : np.nan})
```

```
train_x_x
```

```
train_x_x=train_x_x.astype(float)
train_y=train_y.astype(float)
test_x_x=test_x_x.astype(float)
test_y=test_y.astype(float)
```

```
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(train_x_x)
train_x=scaler.fit_transform(train_x_x)
test_x=scaler.fit_transform(test_x_x)
```

```
train_x=pd.DataFrame(train_x)
test_x=pd.DataFrame(test_x)
```

```
train_x.columns=train_x_x.columns
```

```
test_x.columns=train_x_x.columns
```

```
train_x
```

```
test_x
```

```
train_x=train_x.drop(['가구당주차대수'],axis=1)
test_x=test_x.drop(['가구당주차대수'],axis=1)
```

```
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
regression=LinearRegression()
regression.fit(train_x,train_y['target_west_log'])
train_predictions = regression.predict(train_x)
lin_mse=mean_squared_error(train_y['target_west_log'], train_predictions) #타겟값, 예측값
lin_rmse=np.sqrt(lin_mse)
lin_rmse
```

```
x_y_data=pd.concat([x_data,y_data],axis=1)
x_y_data
```

```
train_train
```

```
test_test
```

```
train_train_x=train_train[['x','y']]
train_train_y=train_train.drop(['x','y'],axis=1)
test_test_x=test_test[['x','y']]
test_test_y=test_test.drop(['x','y'],axis=1)
```

```
z_scaler=StandardScaler()
z_scaler.fit(train_train_x)
train_x_z=z_scaler.transform(train_train_x)
test_x_z=z_scaler.transform(test_test_x)
```

```
train_x_z=pd.DataFrame(train_x_z)
train_x_z.columns=['x','y']
train_x_z
```

```
test_x_z=pd.DataFrame(test_x_z)
test_x_z.columns=['x','y']
test_x_z
```

```
model1=LinearRegression()
model1.fit(train_x_z,train_train_y['target_west_log'])
mse_west_west=mean_squared_error(train_train_y['target_west_log'],model1.predict(train_x_z))
print(np.sqrt(mse_west_west))
print(model1.score(train_x_z,train_train_y['target_west_log']))
```

```
model2=LinearRegression()
model2.fit(train_x_z,train_train_y['target_east_log'])
mse_east_east=mean_squared_error(train_train_y['target_east_log'],model2.predict(train_x_z))
print(np.sqrt(mse_east_east))
print(model2.score(train_x_z,train_train_y['target_east_log']))
```

```
model3=LinearRegression()
model3.fit(train_x_z,train_train_y['target_north_lat'])
mse_north_north=mean_squared_error(train_train_y['target_north_lat'],model3.predict(train_x_z))
print(np.sqrt(mse_north_north))
print(model3.score(train_x_z,train_train_y['target_north_lat']))
```

```
model4=LinearRegression()
model4.fit(train_x_z,train_train_y['target_south_lat'])
mse_south_south=mean_squared_error(train_train_y['target_south_lat'],model4.predict(train_x_z))
print(np.sqrt(mse_south_south))
print(model4.score(train_x_z,train_train_y['target_south_lat']))
```

```
mse_w=mean_squared_error(test_test_y['target_west_log'],model1.predict(test_test_x))
mse_e=mean_squared_error(test_test_y['target_east_log'],model2.predict(test_test_x))
mse_n=mean_squared_error(test_test_y['target_north_lat'],model3.predict(test_test_x))
mse_s=mean_squared_error(test_test_y['target_south_lat'],model4.predict(test_test_x))

print(np.sqrt(mse_w))
print(np.sqrt(mse_e))
print(np.sqrt(mse_n))
print(np.sqrt(mse_s))
```
