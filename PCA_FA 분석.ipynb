# 결측값들을 모두 0으로 대체한다.
여기서 0은 결측값을 하나의 범주로 생각한다는 얘기인데, 이는 다음과 같은 이유로 0으로 대체한다.

1. 데이터의 표본이 적다.

결측값을 제거하기에는 채 300도 안되는 표본수가 나오기 때문에, 데이터의 양을 확보하고 양질의 분석을 하기 위해서 제거하는 방법은 바람직 하지 않다.

2. 결측값 또한 하나의 값이다.

일반적으로 결측값을 대체하는 방법으로는 평균, 중간값, 회귀, 비지도학습 대체 등등 많은 방법이 있지만, 이 데이터의 경우 마찬가지이유로 표본수가 적기때문에 많은 오류를 발생시킬 우려가 있다. 그렇기 때문에 결측값을 하나의 범주로 생각하고, 분석을 진행하면 데이터의 오염을 최소한으로 줄일 수 있다. 또한 연속형 변수들의 경우에는 0으로 대체하는것이 더욱 효과적으로 작용한다.

```python
data=data.fillna(0)
```

# PCA 분석을 위한 범주형 변수들 제거

```python
data=data.reset_index()
data=data.drop(['index','sido','openDate','instkind','sgg','ownerChange','inst_id'],axis=1)
data
```

# 타겟변수변환
타겟변수는 범주형 변수이기 때문에 0,1로 변환하는 인코딩을 진행한다.

```python
from sklearn.preprocessing import OrdinalEncoder
data_y=data['OC']
data_y=pd.DataFrame(data_y)
encoder=OrdinalEncoder()
data_y=encoder.fit_transform(data_y)
data_y=pd.DataFrame(data_y)
data_y.columns=['OC']
data_y
```

```python
data.describe()
```

# 로짓모형을 통한 상관관계 살펴보기

종속변수는 범주형, 독립변수는 연속형 변수이므로 일반적인 상관계수를 통한 분석은 유의하지 않다. 그러므로, 로짓모형을 통해 각 변수들의 p-value를 살펴보고, 각 독립변수들이 종속변수와 상관이 있는가를 살펴보도록 한다.
이때, 귀무가설을 기각하는 p-value는 0.05로 설정한다.

```python
import statsmodels.api as sm
empty=pd.DataFrame()
for ind_x in data_x:
  logit = sm.Logit(data_y,data_x[ind_x].astype(float)) #로지스틱 회귀분석 시행
  result = logit.fit()
  if float(result.pvalues.values) <0.05:
    indra_x=pd.DataFrame([ind_x])
    pval=pd.DataFrame(result.pvalues.values)
    dfdf=pd.concat([indra_x,pval],axis=1)
    empty=pd.concat([empty,dfdf],axis=0)
    print('종속변수와 관계가 있는',ind_x+'의 p-value 값 :',result.pvalues.values)
    dfdf=pd.DataFrame()
```

# 상관이 있다고 결론난 변수들 테이블화

오름차순으로 정렬한다.
상단에 있을수록 귀무가설을 기각하는 강한 반증이 된다.

```python
empty=empty.reset_index()
empty=empty.drop(['index'],axis=1)
empty.columns=['변수','P-values']
empty=empty.sort_values(by=['P-values'])
empty
```

# 가장 유의한 변수(p_value min, p_value max)

```python
print(empty[empty['P-values']==empty['P-values'].min()])
print(empty[empty['P-values']==empty['P-values'].max()])
```

# PCA 및 요인분석을 위한 표준화 스케일링 시작

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
data_scale_x=scaler.fit_transform(data_x)
data_scale_x=pd.DataFrame(data_scale_x)
data_scale_x.columns=data_x.columns
data_scale_x
```

# 주성분 분석 실시

```python
from sklearn.decomposition import PCA
pca=PCA() 
pca.fit_transform(data_scale_x)
```

# 주성분 갯수를 정하기 위한 scree plot 작성
고유값을 이용하여 그래프를 작성해본다.
통상, 그래프가 꺾이는 시점을 주성분의 개수로 정한다.

```python
import matplotlib.pyplot as plt
plt.plot(pca.explained_variance_, '*')
plt.ylabel('explained_variance')
plt.xlabel('variables')
```

# 주성분 축들의 정보량(분산량) 살펴보기
고유값 행렬을 통해 정보량을 살펴본다.

```python
pca.explained_variance_
```

# 모든 고유값들을 더하면 변수의 개수가 되어야한다.
변수의 개수 = 49

```python
np.cumsum(pca.explained_variance_)
```

# 주성분 분산 비율을 확인한다.
이 행렬을 통해서 각 주성분 축이 갖고있는 정보의 양(분산량)을 확인할 수 있다.

```python
pca.explained_variance_ratio_
```

# 마찬가지로 모든 정보량의 합은 1이 되어야한다.

```python
np.cumsum(pca.explained_variance_ratio_)
```

# 주성분 축의 고유 벡터 행렬 확인

```python
vector_mat=pd.DataFrame(pca.components_)
vector_mat
```

# 주성분 개수 정하기
scree plot의 경우에는 변수의 수가 많아서 정확한 확인이 어렵다. 또한, 고유값 행렬에서는 1 이상이 넘는 것들로만 모아도 많은 수준의 정보량을 확보할 수 있기에 70% 정보량을 설명할 수 있는 주성분의 갯수를 계산해본다.

이 연구의 목적은 최대한의 정보량을 가진 축을 골라내는 작업을 통해 차원축소를 하는것에 있는것이 아니기 때문에 최대한의 정보량을 확보하기 위하여 주성분 축의 갯수를 늘릴 필요는 없다. 또한, 주성분 축의 갯수가 늘어나게 되면 오히려 해석하는데에 있어서 혼동이 올 수 있고, 분산량이 비슷한 경우 주성분 축끼리의 경계가 모호해질 수 있다.

```python
print('70% 이상의 정보량을 갖고 있는 주성분 갯수 :',np.argmax(np.cumsum(pca.explained_variance_ratio_)>=0.7)+1)
```

# 주성분 축 4개로 지정하여 다시 시작
주성분 축 4개로 설명할 수 있는 분산량은 약 75%이다.

```python
pca_4=PCA(n_components=4) 
pca_4_result=pd.DataFrame(pca_4.fit_transform(data_scale_x))
pca_4_result
```

# 고유벡터 확인

첫 번째 주성분 부터 4번째 까지 주성분 축을 분석해 본다.

첫 번째 주성분의 경우는 상단에 위치한 변수들이 서로 비슷한 고유 벡터값을 갖고 있다. 이는 곧, 판관비, 매출액, 부채총계, 비유동자산, 봉급, 당좌자산 등 전반적인 재무상태에 관련한 것을 나타낸다.

```python
pca_component=pd.DataFrame(pca_4.components_)
pca_component.columns=data_x.columns
pca_component=pca_component.T
pca_component=pca_component.sort_values(by=[0],ascending=False)
pca_component
```

# 두 번째 주성분
세금과 수익에 관련한 부분이 +축에, 장기차입금, 이자비용 등은 -축에 위치해 있다. 이는 곧 두 번째 주성분 축은 재무상태의 건실함 정도를 나타낸다고 할 수 있다.

```python
pca_component=pca_component.sort_values(by=[1],ascending=False)
pca_component
```

# 세 번째 주성분
매출원가, 기타 비유동자산은 +방향에,
이익잉여금, 당기순이익등은 -방향에 놓여있다. 이는 곧, 재무상태의 악화정도를 나타낸다고 볼 수 있다.

```python
pca_component=pca_component.sort_values(by=[2],ascending=False)
pca_component
```

# 네 번째 주성분
단기미수금, 순자산등이 + 방향에, 
당기순이익과 기타비유동자산이 - 방향에 위치해 있다.
이는 곧, 순자산의 규모에 관한 것임을 알 수 있다.

```python
pca_component=pca_component.sort_values(by=[3],ascending=False)
pca_component
```

# 주성분법을 이용한 인자분석을 활용한다.
