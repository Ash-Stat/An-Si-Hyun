```
import pandas as pd
o=pd.Series([36.411958,36.412195,36.412466,36.412545,36.412783,36.412240])
a=pd.Series([127.393285,127.382247,127.382285,127.382321,127.393995,127.382258])    
o=pd.DataFrame(o)
a=pd.DataFrame(a)
a=pd.concat([a,o],axis=1)
a.columns=['longitude','latitude']
a
```

```
import cv2
import matplotlib.pyplot as plt
import numpy as np
for x,y in zip(a['longitude'],a['latitude']):
  name=str(y)+' '+str(x)+'.jpg'
 
  img=cv2.imread(name,cv2.IMREAD_COLOR)
  
  plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
  plt.show()
  
```
