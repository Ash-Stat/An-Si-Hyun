```
import pandas as pd
import numpy as np
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
import tensorflow as tf
from tensorflow import keras
```

# 데이터셋 불러오기


```
train=pd.read_csv('./train.csv')
test=pd.read_csv('./test.csv')
```

# 결측값 확인하기

```
train.isnull().sum()
```

```
# 데이터 정리는 내가 해놨던거 참고
train_x=train
```

# 날릴거 날리고 채울거 채우기

```
train_x=train_x.drop(['Ticket','Name','Cabin','PassengerId','Parch'],axis=1)
train_x
```

```
mean_age=train_x['Age'].fillna(train_x['Age'].mean())
median_age=train_x['Age'].fillna(train_x['Age'].median())
```

```
train_mean_x=train_x
train_median_x=train_x
```

```
train_mean_x['Age']=mean_age
train_mean_x=train_mean_x.dropna()
train_mean_x=train_mean_x.reset_index()
train_mean_x=train_mean_x.drop(['index'],axis=1)
```

```
train_median_x['Age']=median_age
train_median_x=train_median_x.dropna()
train_median_x=train_median_x.reset_index()
train_median_x=train_median_x.drop(['index'],axis=1)
```

```
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()
category=encoder.fit_transform(train_mean_x[['Sex','Embarked']])
category=pd.DataFrame(category)
category.columns=['Sex','Embarked']
category
```

```
encoder = OrdinalEncoder()
category1=encoder.fit_transform(train_median_x[['Sex','Embarked']])
category1=pd.DataFrame(category1)
category1.columns=['Sex','Embarked']
category1
```

```
train_mean_x['Sex']=category['Sex']
train_mean_x['Embarked']=category['Embarked']

train_median_x['Sex'] = category1['Sex']
train_median_x['Embarked'] = category1['Embarked']
```

```
train_mean_x
```

```
train_median_x
```

```
from sklearn.preprocessing import MinMaxScaler

Scaler = MinMaxScaler()
train_mean_y=train_mean_x['Survived']
train_mean_x=train_mean_x.drop(['Survived'],axis=1)
mean_scaler=Scaler.fit_transform(train_mean_x)
mean_scaler=pd.DataFrame(mean_scaler)
mean_scaler.columns=['Pclass','Sex','Age','SibSp','Fare','Embarked']
mean_scaler
```

```
Scaler = MinMaxScaler()
train_median_y=train_median_x['Survived']
train_median_x=train_median_x.drop(['Survived'],axis=1)
median_scaler=Scaler.fit_transform(train_median_x)
median_scaler=pd.DataFrame(median_scaler)
median_scaler.columns=['Pclass','Sex','Age','SibSp','Fare','Embarked']
median_scaler
```

```
train_mean_x,valid_mean_x,train_mean_y,valid_mean_y=train_test_split(mean_scaler,train_mean_y,test_size=0.2,random_state=42)
i_input=keras.layers.Input(shape=train_mean_x.shape[1:])
hidden1=keras.layers.Dense(16, activation='relu')(i_input)
hidden2=keras.layers.Dense(32, activation='relu')(hidden1)
hidden3=keras.layers.Dense(64, activation='relu')(hidden2)
hidden4=keras.layers.Dense(128, activation='relu')(hidden3)
hidden5=keras.layers.Dense(128, activation='relu')(hidden4)
i_concat=keras.layers.Concatenate()([i_input,hidden5])
i_output=keras.layers.Dense(1,activation='sigmoid')(i_concat)
i_model=keras.Model(inputs=[i_input], outputs=[i_output])
adam=keras.optimizers.Adamax(lr=0.03)
i_model.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])
i_history = i_model.fit(train_mean_x, train_mean_y, epochs=1000, 
                   batch_size=20,callbacks=[keras.callbacks.EarlyStopping(patience=100, 
                   restore_best_weights=True)], validation_data=(valid_mean_x,valid_mean_y), verbose=1)
DNN_evaluation=i_model.evaluate(valid_mean_x,valid_mean_y)
```

```
train_median_x,valid_median_x,train_median_y,valid_median_y=train_test_split(median_scaler,train_median_y,test_size=0.2,random_state=42)
i_input=keras.layers.Input(shape=train_median_x.shape[1:])
hidden1=keras.layers.Dense(64, activation='relu')(i_input)
hidden2=keras.layers.Dense(64, activation='relu')(hidden1)
hidden3=keras.layers.Dense(64, activation='relu')(hidden2)
hidden4=keras.layers.Dense(64, activation='relu')(hidden3)
hidden5=keras.layers.Dense(64, activation='relu')(hidden4)
i_concat=keras.layers.Concatenate()([i_input,hidden5])
i_output=keras.layers.Dense(1,activation='sigmoid')(i_concat)
i_model=keras.Model(inputs=[i_input], outputs=[i_output])
adam=keras.optimizers.Adamax(lr=0.03)
i_model.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])
i_history = i_model.fit(train_median_x, train_median_y, epochs=1000, 
                   batch_size=20,callbacks=[keras.callbacks.EarlyStopping(patience=100, 
                   restore_best_weights=True)], validation_data=(valid_median_x,valid_median_y), verbose=1)
DNN_evaluation=i_model.evaluate(valid_median_x,valid_median_y)
```
