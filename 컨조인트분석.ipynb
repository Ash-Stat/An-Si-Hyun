```
import pandas as pd
s=pd.read_csv('./답변데이터.csv')

#파이썬으로 데이터 불러오기
```

```
s=s.drop(['Unnamed: 6'],axis=1)
s

#필요없는 칼럼 제거
```

```
sunho=s['선호도']
s=s.drop(['선호도'],axis=1)
dum1=pd.get_dummies(s['위협문구'])
dum2=pd.get_dummies(s['글자크기'])
dum3=pd.get_dummies(s['경고그림'])
dum4=pd.get_dummies(s['박스크기'])
dummies=pd.concat([s,dum1,dum2,dum3,dum4,sunho],axis=1)
dummies

#각각 속성에대한 더미변수 생성
```

```
dummies.to_csv('./더미변수.csv',encoding='utf-8-sig')

#더미변수생성한 데이터셋을 저장
```

```
dum=pd.read_csv('./덤덤덤덤.csv')
y=dum['선호도']
dum=dum.drop(['Unnamed: 0','카드','선호도'],axis=1)

#컨조인트 분석을 위한 데이터셋 가공
```

```
dum=dum.rename({'위협문구':'Threat','글자크기':'FontSize','경고그림':'Warning','박스크기':'BoxSize'},axis='columns')
dum

#컨조인트 데이터셋을 위한 데이터 전처리
```

```
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('bmh')


regression = sm.OLS(y, dum, family=sm.families.Binomial()).fit()
regression.summary()
#컨조인트 분석 실시 및 결과
```

```
import numpy as np
regree = pd.DataFrame({
    'param_name': regression.params.keys()
    , 'param_w': regression.params.values
    , 'pval': regression.pvalues
})
# adding field for absolute of parameters
regree['abs_param_w'] = np.abs(regree['param_w'])
# marking field is significant under 95% confidence interval
regree['is_sig_95'] = (regree['pval'] < 0.05)
# constructing color naming for each param
regree['c'] = ['blue' if x else 'red' for x in regree['is_sig_95']]

# make it sorted by abs of parameter value
regree = regree.sort_values(by='abs_param_w', ascending=True)
f, ax = plt.subplots(figsize=(14, 8))
plt.title('Part Worth')
pwu = regree['param_w']
xbar = np.arange(len(pwu))
plt.barh(xbar, pwu, color=regree['c'])
plt.yticks(xbar, labels=regree['param_name'])
plt.show()

#p-value 시각화
```

```
def utility(a,b,c,d,e,f,g,h):
  uty=(-3.2526*a)+(-3.5317*b)+(-3.2991)*c+(-2.3754*d)+(-3.8899*e)+(-3.5014*f)+(-4.2590*g)+(-2.6887*h)
  return uty

  #효용 계산 함수
```

```
print('시나리오 첫번째 : ',utility(0,0,0,1,0,1,0,1)) # 3,2,2,2
print('시나리오 두번째 : ',utility(0,1,0,1,0,0,0,1)) #2,2,3,2
print('시나리오 세번째 : ',utility(1,0,0,0,1,0,0,0)) #1,3,1,3
print('시나리오 네번째 : ',utility(0,0,1,0,0,1,1,0)) #3,1,2,1

#시나리오별 효용도 계산
```
