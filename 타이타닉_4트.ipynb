```
import pandas as pd
import numpy as np
```

```
train=pd.read_csv('./train.csv')
test=pd.read_csv('./test.csv')
submission=pd.read_csv('./gender_submission.csv')
```

```
train=train.drop(['Cabin','Name','Ticket','PassengerId'],axis=1)
train
```

```
empty=pd.DataFrame()
for a,b in zip(train['Embarked'].isnull(), train.index):
  if a == False:
    ss=train.loc[b]
    ss=pd.DataFrame(ss)
    ss=ss.T
    empty=pd.concat([empty,ss],axis=0)
empty=empty.reset_index()
empty=empty.drop(['index'],axis=1)
empty
```

```
from sklearn.preprocessing import OrdinalEncoder
encoder=OrdinalEncoder()
enco=encoder.fit_transform(empty[['Sex','Embarked']])
enco=pd.DataFrame(enco)
enco.columns=['Sex','Embarked']
empty['Sex']=enco['Sex']
empty['Embarked'] = enco['Embarked']
empty
```

```
import statsmodels.api as sm
data=empty[['Parch','Survived']]
data=data.astype(float)
logit=sm.Logit(data['Survived'],data['Parch'])
result=logit.fit()
result.summary2()
```
