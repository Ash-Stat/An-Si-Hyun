```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import cv2
from cv2 import imread, imshow, IMREAD_COLOR, COLOR_BGR2RGB, cvtColor
import matplotlib.image as mpimg

lower_orange = (100, 100, 100)
upper_orange = (255, 255, 255)
lower_green = (50, 50, 50)
upper_green = (200, 200, 200)
lower_blue = (0, 180, 55)
upper_blue = (20, 255, 200)
Daejeon = pd.read_csv('./why.csv', encoding='cp949')
longitude = Daejeon['x']
latitude = Daejeon['y']
result = pd.DataFrame()
result2 = pd.DataFrame()
result3 = pd.DataFrame()
for a in np.arange(127.301355,127.470424,0.000001):
    for b in np.arange(36.28276,36.453431,0.000001):
        for log, lat in zip(longitude, latitude):
            if b == lat and a == log :
                print(a,b)
                name = str(b) + ' ' + str(a) + '.jpg'
                img = mpimg.imread(name, cv2.IMREAD_COLOR)
                img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
                img_mask = cv2.inRange(img_hsv, lower_green, upper_green)
                img_result = cv2.bitwise_and(img, img, mask=img_mask)
                log_lat = pd.DataFrame([[a,b]])

                result = pd.concat([result, log_lat], axis=0)

                for x in range(0, 416):
                    for y in range(0, 832):
                        if img_result[x, y][0] and img_result[x, y][1] and img_result[x, y][2] != 0:
                            r = pd.DataFrame([x])
                            result2 = pd.concat([result2, r], axis=0)
                            print('높이 좌표 :', x, '', '넓이 좌표 :', y)
                leng = len(result2)
                length = (pd.DataFrame([leng])/(832*416))*100
                result3 = pd.concat([result3, length], axis=0)

print(result)
print(result3)
result=pd.concat([result,result3],axis=1)
print(result)
```
