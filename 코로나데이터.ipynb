```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
Patientinfo=pd.read_csv('./PatientInfo.csv')
Patientinfo
```

```
Patientinfo['count']=1
```

```
Patientinfo
```

```
P_M=Patientinfo[Patientinfo['sex'] == 'male'][['count','confirmed_date']].groupby('confirmed_date')
P_W=Patientinfo[Patientinfo['sex'] == 'female'][['count','confirmed_date']].groupby('confirmed_date')
```

```
P_M.sum()
P_M=pd.DataFrame(P_M.sum())
P_M=P_M.reset_index()
```

```
P_W.sum()
P_W=pd.DataFrame(P_W.sum())
P_W=P_W.reset_index()
```

```
plt.figure(figsize=(15,15))
plt.xticks(rotation=-90)
plt.title('Patient who is man')
sns.lineplot(data=P_M, x='confirmed_date',y='count',linestyle=':',markers='o')
plt.show()
```

```
plt.figure(figsize=(15,15))
plt.xticks(rotation=-90)
plt.title('Patient Datetime Graph')
sns.lineplot(data=P_M, x='confirmed_date',y='count',linestyle=':',markers='o')
sns.lineplot(data=P_W, x='confirmed_date',y='count',linestyle='--',markers='o')
```
