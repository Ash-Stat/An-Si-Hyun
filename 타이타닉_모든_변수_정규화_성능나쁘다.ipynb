```
import pandas as pd
import numpy as np
```

# 데이터셋 불러오기


```
train=pd.read_csv('./train.csv')
test=pd.read_csv('./test.csv')
```

# 결측값 확인하기

```
train.isnull().sum()
```

```
# 데이터 정리는 내가 해놨던거 참고
train_x=train
```

# 날릴거 날리고 채울거 채우기

```
train_x=train_x.drop(['Ticket','Name','Cabin','PassengerId','Parch'],axis=1)
train_x
```

```
mean_age=train_x['Age'].fillna(train_x['Age'].mean())
median_age=train_x['Age'].fillna(train_x['Age'].median())
```

```
train_mean_x=train_x
train_median_x=train_x
```

```
train_mean_x['Age']=mean_age
train_mean_x=train_mean_x.dropna()
train_mean_x=train_mean_x.reset_index()
train_mean_x=train_mean_x.drop(['index'],axis=1)
```

```
train_median_x['Age']=median_age
train_median_x=train_median_x.dropna()
train_median_x=train_median_x.reset_index()
train_median_x=train_median_x.drop(['index'],axis=1)
```

```
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()
category=encoder.fit_transform(train_mean_x[['Sex','Embarked']])
category=pd.DataFrame(category)
category.columns=['Sex','Embarked']
category
```

```
encoder = OrdinalEncoder()
category1=encoder.fit_transform(train_median_x[['Sex','Embarked']])
category1=pd.DataFrame(category1)
category1.columns=['Sex','Embarked']
category1
```

```
train_mean_x['Sex']=category['Sex']
train_mean_x['Embarked']=category['Embarked']

train_median_x['Sex'] = category1['Sex']
train_median_x['Embarked'] = category1['Embarked']
```

```
train_mean_x
```

```
train_median_x
```

```
from sklearn.preprocessing import MinMaxScaler
train_mean_y=train_mean_x['Survived']
train_mean_x=train_mean_x.drop(['Survived'],axis=1)
Scaler = MinMaxScaler()

mean_scaler=Scaler.fit_transform(train_mean_x)
mean_scaler=pd.DataFrame(mean_scaler)
mean_scaler.columns=['Pclass','Sex','Age','SibSp','Fare','Embarked']
mean_scaler
```

```
train_median_y=train_median_x['Survived']
train_median_x=train_median_x.drop(['Survived'],axis=1)
Scaler = MinMaxScaler()

median_scaler=Scaler.fit_transform(train_median_x)
median_scaler=pd.DataFrame(median_scaler)
median_scaler.columns=['Pclass','Sex','Age','SibSp','Fare','Embarked']
median_scaler
```

```
# 트레인이랑 테스트 데이터셋 나누진 않는다. 왜냐하면
# 어차피 테스트 트레인이 나누어져 있기도하고, 데이터 수가 적기 때문에
```

```
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
RF= RandomForestClassifier()
parameter=[
           {'n_estimators' : [100,150,200,250], 'max_features' : [1,2,3,4,5,6]},
           {'bootstrap' : [False], 'n_estimators' : [100,150,200,250], 'max_features' : [1,2,3,4,5,6]}
]
grid_mean=GridSearchCV(RF,parameter,verbose=2,return_train_score=True,scoring='accuracy',cv=5)
grid_mean.fit(mean_scaler,train_mean_y)
mean_best_forest=grid_mean.best_estimator_
```

```
RF= RandomForestClassifier()
parameter=[
           {'n_estimators' : [100,150,200,250], 'max_features' : [1,2,3,4,5,6]},
           {'bootstrap' : [False], 'n_estimators' : [100,150,200,250], 'max_features' : [1,2,3,4,5,6]}
]
grid_median=GridSearchCV(RF,parameter,verbose=2,return_train_score=True,scoring='accuracy',cv=5)
grid_median.fit(median_scaler,train_median_y)
median_best_forest=grid_median.best_estimator_
```

```
test=pd.read_csv('./test.csv')
test
```

```
test=test.drop(['PassengerId','Name','Parch','Ticket','Cabin'],axis=1)
test
```

```
encoder = OrdinalEncoder()
category2=encoder.fit_transform(test[['Sex','Embarked']])
category2=pd.DataFrame(category2)
category2.columns=['Sex','Embarked']
category2
test['Sex']=category2['Sex']
test['Embarked'] = category2['Embarked']

test_mean=test
test_median=test
test_mean['Age']=test_mean['Age'].fillna(test_mean['Age'].mean())
test_mean['Fare']=test_mean['Fare'].fillna(test_mean['Fare'].mean())
test_median['Age']=test_median['Age'].fillna(test_median['Age'].median())
test_median['Fare']=test_median['Fare'].fillna(test_median['Fare'].median())
```

```
Scaler = MinMaxScaler()

mean_scaler=Scaler.fit_transform(test_mean)
mean_scaler=pd.DataFrame(mean_scaler)
mean_scaler.columns=['Pclass','Sex','Age','SibSp','Fare','Embarked']
mean_scaler
```

```
Scaler = MinMaxScaler()

median_scaler=Scaler.fit_transform(test_median)
median_scaler=pd.DataFrame(median_scaler)
median_scaler.columns=['Pclass','Sex','Age','SibSp','Fare','Embarked']
median_scaler
```

```
mean_predict=mean_best_forest.predict(mean_scaler)
median_predict=median_best_forest.predict(median_scaler)
```

```
submission_mean=pd.read_csv('./gender_submission.csv')
submission_median=pd.read_csv('./gender_submission.csv')
```

```
submission_mean['Survived']=mean_predict
submission_median['Survived']=median_predict
```

```
submission_mean.to_csv('./mean_scaler.csv',encoding='utf-8-sig',index=False)
submission_median.to_csv('./median_scaler.csv',encoding='utf-8-sig',index=False)
```
