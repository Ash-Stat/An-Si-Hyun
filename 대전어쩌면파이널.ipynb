```
import pandas as pd
import numpy as np
```

```
#스트리트뷰 이미지 다운로드를 위해서 파노라마 분리
panorama=pd.read_csv('./대전파노라마.txt',sep=',')
coordinates=pd.read_csv('./대전좌표.txt',sep=',')
```

```
panorama
```

```
panorama_download=panorama['파노라마']
panorama_download
```

# txt로 저장하면 #NAME? 이렇게 안떠서 바로 복사가능

```
panorama_download.to_csv('./panoarama_daejoen.txt',encoding='utf-8-sig',index=False)
```

```
#다운받는 파노라마랑 데이터로 저장한 파노라마랑 값의 수가 일치하는지 확인
panorama_download=pd.DataFrame([panorama_download])
panorama_download=panorama_download.T
panorama_download
```

```
#5802로 일치~~
panorama_download_up=panorama_download.drop_duplicates()
panorama_download_up
```

```
#파이참에서 녹색지수 추출할 데이터셋 가공 후 불러오기
green_result=pd.read_csv('./DaejeonGreenIndex.csv')
green_result
```

```
#파노라마에 그린인덱스 맵핑시켜주기 #고유파라노마데이터셋에

panorama_download_up['HGVI']=np.nan
panorama_download_up['파노라마log']=np.nan
panorama_download_up['파노라마lat']=np.nan
panorama_download_up.index=panorama_download_up['파노라마']
```

```
for pa,hgvi,lo,la in zip(green_result['2'],green_result['0.1'],green_result['0'],green_result['1']):
  pa=pa[:-4]
  panorama_download_up['HGVI'].loc[pa] =  hgvi
  panorama_download_up['파노라마log'].loc[pa] =  lo
  panorama_download_up['파노라마lat'].loc[pa] =  la
panorama_download_up
```

```
coordinates
```

```
#데이터셋 합쳐주기
coor_pa=pd.concat([coordinates,panorama_download],axis=1)
coor_pa
```

```
panorama_download_up
```

```
coor_pa.index=coor_pa['파노라마']
coor_pa
```

```
coor_pa['파노라마log']=np.nan
coor_pa['파노라마lat']=np.nan
coor_pa['HGVI']=np.nan
for paa,hgvia,loa,laa in zip(panorama_download_up['파노라마'],panorama_download_up['HGVI'],panorama_download_up['파노라마log'],panorama_download_up['파노라마lat']):
  coor_pa['파노라마log'].loc[paa]=loa
  coor_pa['파노라마lat'].loc[paa]=laa
  coor_pa['HGVI'].loc[paa]=hgvia
coor_pa
```

```
coor_pa.index=coor_pa['좌표']
coor_pa
```

```
#locationn을 불러와서 실제 데이터와의 좌표를 맵핑시킨다.
location=pd.read_csv('./locationn.csv')
location.index=np.arange(1,763)
location
```

```
for index_val,x in zip(location.index,location['x']):
  coor_pa['좌표'].loc[index_val]=x
coor_pa
```

```
#파노라마 중복된거 없애주기
coor_pa_dup=coor_pa.drop_duplicates('파노라마')
coor_pa_dup
```

```
len(coor_pa_dup['좌표'].unique())
```

```
final_index=coor_pa_dup[['파노라마log','파노라마lat','HGVI','log','lat']].groupby(coor_pa_dup['좌표']).mean()
final_index
```

```
final_index=final_index.reset_index()
final_index
```

```
# 0이 생략되는 경우 방지

for coordinat,ran in zip(final_index['좌표'],np.arange(0,749)):
  coordinat = format(coordinat,'.6f')
  final_index['좌표'].loc[ran]=coordinat
final_index
```

```
Daejeon=pd.read_excel('./Daejeon_merge_2018.xlsx')
Daejeon['HGVI']= np.nan
Daejeon['d_log']= np.nan
Daejeon['d_lat']= np.nan
Daejeon
```

```
for d_a,d_index in zip(Daejeon['x'],Daejeon.index):
  d_a=format(d_a,'.6f')
  for t_c,t_log,t_lat,t_hgvi in zip(final_index['좌표'],final_index['파노라마log'],final_index['파노라마lat'],final_index['HGVI']):
    if d_a == t_c:
      Daejeon['HGVI'].loc[d_index] = t_hgvi
      Daejeon['d_log'].loc[d_index] = t_log 
      Daejeon['d_lat'].loc[d_index] = t_lat
Daejeon
```

```
Daejeon['error_x']=Daejeon['x']-Daejeon['d_log']
Daejeon['error_y']=Daejeon['y']-Daejeon['d_lat']
Daejeon['error_y']=abs(Daejeon['error_y'])
Daejeon['error_x']=abs(Daejeon['error_x'])
Daejeon
```

```
Daejeon.to_csv('./Daejeon_HGVI.csv',encoding='utf-8-sig')
```

```
print('경도 평균 오차 :' ,Daejeon['error_x'].mean())
print('위도 평균 오차 :' ,Daejeon['error_y'].mean())
```
