```
import cv2 #Open CV 호출
import matplotlib.pyplot as plt
#cv2.imread(file_name, flag) #이미지를 읽어서 numpy 객체로 만드는 함수
#flag : IMREAD_COLOR = 이미지를 Color로 읽고 투명한 부분은 무시
#flag : IMREAD_GRAYSCALE = 이미지를 Grayscale로 읽기
#flag : IMREAD_UNCHANGEED : 이미지를 Color로 읽고, 투명한 부분도 읽기(Alpha)

#반환 값:Numpy  객체(행,열,색상 : 기본 BGR)

#cv2.imshow(title,image) 특정한 이미지를 화면에 출력
#title : 윈도우 창의 제목
#image : 출력할 이미지 객체

#cv2.imwrite(file_name,image) : 특정한 이미지를 파일로 저장하는 함수
#file_name : 저장할 이미지 파일 이름
#image : 저장할 이미지 객체

#cv2.waitKey(time) : 키보드 입력을 처리하는 함수
#time = 입력 대기 시간(무한대기 : 0, 아무키나 누르면 그때 처리)
#반환 값: 사용자가 입력한 Ascii Code(ESC : 27)

#cv2.destoryAllWindows() : 화면의 모든 윈도우를 닫는 함수
#파이참의 경우 쓸만하다
```

```
img=cv2.imread('ima.jpg',cv2.IMREAD_COLOR)
#cv2.imshow('Kaist',img) #코랩에선 이렇게하면 안된다.

img_gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #cvtColor를 통해 읽어 들였던 이미지를 BGR그레이로 바꿀 수 있다.
cv2.imwrite('result16.png',img)
cv2.imwrite('result16_gray.png',img_gray)
```

```
#코랩으로 이미지 출력하기

plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB)) #plt는 RGB를 기준으로 한다는 중요한 차이가 있다. 따라서 BGR to RGB를 넣어주어야 한다.
plt.show()

plt.imshow(cv2.cvtColor(img_gray,cv2.COLOR_GRAY2RGB)) #마찬가지로 그레이로 바꾼다음에 RGB로 바꾸어 주어야 한다.
plt.show()
```

```
#이미지 크기 및 픽셀 확인

print(img.shape) #높이, 너비
print(img.size)

#이미지 Numpy 객체의 특정 픽셀을 가리키는 것
px = img[100,100]
print(px)

#B,G,R 순서로 출력 된다.
#단, Gray Sclae인 경우에는 B,G,R로 구분되지 않는다.
```

```
#이미지 특정 범위 픽셀 변경
import time

start_time=time.time()
for i in range(0,100):
  for j in range(0,100):
    img[i,j] = [255,255,255]
print('---%s seconds ---' %(time.time() - start_time))
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()
#슬라이싱을 이용한 범위 픽셀 변경
strat_time=time.time()
img[0:100,0:100]=[255,255,255]
print('---%s seconds ---' %(time.time() - strat_time))
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()
```

```
#특정한 이미지 픽셀 처리
img=cv2.imread('ima.jpg',cv2.IMREAD_COLOR)
img[:,:,2]=50 #맨마지막에 오는 숫자 BGR순으로 픽셀값 바꿔준다.
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()
```
